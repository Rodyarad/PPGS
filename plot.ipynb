```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

s1 = pd.Series(oneshot, index=margins)
s2 = pd.Series(full, index=margins)

df = pd.DataFrame({
    'margin': list(s1.index) * 2,
    'path found, %': list(s1) + list(s2),
    'series': ['One-Shot Planner'] * len(s1) + ['Full Planner'] * len(s2)
})

plt.figure(figsize=(8,5))
ax = sns.barplot(df, x='margin', y='path found, %', hue='series')

hline = ax.axhline(y=base_acc, color='red', linestyle='--', linewidth=2, label='Latent BFS')
handles, labels = ax.get_legend_handles_labels()
handles.append(hline)
# labels.append('Latent BFS')
ax.legend(handles=handles, labels=labels)

ax.set(ylim=(0, 1))
plt.show()
```
